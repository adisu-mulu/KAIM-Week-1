```python
# loading libraries
import pandas as pd
```

```python
# loading data 
aapl = pd.read_csv('../Data/yfinance_data/AAPL_historical_data.csv')
amzn = pd.read_csv('../Data/yfinance_data/AMZN_historical_data.csv')
google = pd.read_csv('../Data/yfinance_data/GOOG_historical_data.csv')
meta = pd.read_csv('../Data/yfinance_data/META_historical_data.csv')
msft = pd.read_csv('../Data/yfinance_data/MSFT_historical_data.csv')
nvda = pd.read_csv('../Data/yfinance_data/NVDA_historical_data.csv')
tsla = pd.read_csv('../Data/yfinance_data/TSLA_historical_data.csv')
```

```python
# changing date to datetime 
aapl['Date'] = pd.to_datetime(aapl['Date'])
amzn['Date'] = pd.to_datetime(amzn['Date'])
google['Date'] = pd.to_datetime(google['Date'])
meta['Date'] = pd.to_datetime(meta['Date'])
msft['Date'] = pd.to_datetime(msft['Date'])
nvda['Date'] = pd.to_datetime(nvda['Date'])
tsla['Date'] = pd.to_datetime(tsla['Date'])
```

```python
# Line chart of stock closing price (aapl)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(aapl['Close'], label='Close Price')
plt.title('Stock Closing Prices Over Time - AAPL')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Stock prices of aapl with moving averages
plt.figure(figsize=(10, 5))
plt.plot(aapl['Close'], label='Close Price')
plt.plot(aapl['SMA'], label='20-Day SMA', color='orange')
plt.plot(aapl['EMA'], label='20-Day EMA', color='magenta')
plt.title('Stock Prices with Moving Averages - AAPL')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# RSI of aapl
plt.figure(figsize=(10, 5))
plt.plot(aapl['RSI'], label='RSI', color='purple')
plt.axhline(70, linestyle='--', alpha=0.5, color='red')  # Overbought threshold
plt.axhline(30, linestyle='--', alpha=0.5, color='green')  # Oversold threshold
plt.title('Relative Strength Index (RSI) - AAPL')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.show()
```

```python
#MACD of aapl is plotted along with its signal line to identify buy/sell signals based on crossovers.
plt.figure(figsize=(10, 5))
plt.plot(aapl['MACD'], label='MACD', color='blue')
plt.plot(aapl['MACD_Signal'], label='Signal Line', color='red')
plt.title('MACD and Signal Line - AAPL')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.legend()
plt.show()
```
